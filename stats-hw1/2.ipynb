```python

def get_estimator(n: int, k: int):
    # print(np.power(array[:, :n], k))
    k_mean = (1/n)*(np.power(array[:, :n], k)).sum(axis=1)
    result = np.power(factorial(k)/k_mean, (1/k))
    return result

def get_ys(k: int): # возвращает массив кортежей с игриками для каждого икса
    return np.array([np.linalg.norm(get_estimator(n, k) - 1 / teta) for n in range(1, randvar_count+1)])
        
randvar_count = 10000
sample_count = 1
teta = 1
plt.figure(figsize=(20, 20))

array = sps.expon.rvs(size=(sample_count, randvar_count), scale=teta)

def exec_experiment(k: int):
    X = np.arange(1, randvar_count + 1)
    Y = get_ys(k)
        
    plt.plot(X, Y, alpha=1, label='$\\left(%s! / \\overline{X^{%s}}\\right)^{1/%s}$' % (k, k, k))   
    plt.legend(fontsize=15)
    plt.title('teta = %s' % k)
    plt.ylabel = 'Estimator value'
    plt.xlabel = 'n'
    
for k in range(1, 149, 10):
    exec_experiment(k)

plt.show()
```
